## Now, Here comes some Basic Questions.

1. Which team won the most matches?

```python
match_wins = df['match_winner'].value_counts()
match_wins
```

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y = match_wins.index,x = match_wins.values,palette='viridis')
plt.title("Most match win by team")
```

2. Toss Decision Trends?

```python
sns.countplot(x = df['toss_decision'],palette= 'rainbow')
plt.title("Toss Decision Trends")
```

3. Toss Winner vs Match Winner?

```python
count = df[df['toss-winner'] == df['match_winner']]['match_id'].count()
percentage = (count *100)/df.shape[0]
percentage.round(2)
```

4. How do teams win? (Runs vs Wickets).

```python
sns.countplot(x = df['won_by'],palette= 'rainbow')
plt.title("Won BY")
```

## Key Player Performances

1. Most "Player of the Match" Awards?

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x = count.values , y = count.index, palette= 'mako')
plt.title("Top 10 Player Man of The Match")
```

2. Top Scorers?

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(5)
high
```

```python
high.plot(kind = 'barh')
```

3. 10 Best Bowling Figures?

```python
df['best_bowling_figure'] = df['best_bowling_figure'].fillna('0/0').astype(str)
df['highest_wickets'] = pd.to_numeric(df['best_bowling_figure'].str.extract(r'(\d+)')[0], errors='coerce').fillna(0).astype(int)
df['best_bowling'] = df['best_bowling'].astype(str).str.strip()

top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind='barh')
```

## IPL 2025 ka winner team with logo

```python
winner_team = df['match_winner'].value_counts().idxmax()
winner_team
```

```python
import os
from IPython.display import Image, display

winner = df['match_winner'].value_counts().idxmax().strip()
folder = "IPL Teams Logo"

candidates = [os.path.join(folder, f"{winner}.jpg"), os.path.join(folder, f"{winner}.jpeg")]
logo = next((p for p in candidates if os.path.exists(p)), None)
if logo is None:
    words = [w.lower() for w in winner.split()]
    for f in os.listdir(folder):
        name = f.lower()
        if all(w in name for w in words):
            logo = os.path.join(folder, f); break

print("IPL 2025 Winner:", winner)
if logo and os.path.exists(logo):
    display(Image(filename=logo))
else:
    print("Logo not found in folder:", folder)
```
